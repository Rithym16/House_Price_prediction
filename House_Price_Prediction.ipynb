```python
# Convert the 'ocean_proximity' column to numerical data using one-hot encoding.
houseP = pd.get_dummies(houseP, columns=['ocean_proximity'])

# Now, you should be able to create the heatmap without the error.
sns.heatmap(houseP.corr(), annot=True)
```

```python
X=houseP.drop('median_house_value',axis=1)
y=houseP['median_house_value']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.impute import SimpleImputer

# Define the imputer: you can choose strategy='mean', 'median', or 'most_frequent'
imputer = SimpleImputer(strategy='mean')

# Fit and transform the data to fill in missing values
X = imputer.fit_transform(X)

# Now, re-split the data (if required)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)

# Fit the model again
lr = LinearRegression()
lr.fit(X_train, y_train)

# Make predictions
predictions = lr.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50);
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# R-squared
r2 = r2_score(y_test, predictions)
print("R-squared:", r2)

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, predictions))
print("RMSE:", rmse)
```
